
# RNN
As Redes Neurais Recorrentes (RNNs) são uma classe poderosa de redes neurais que são especialmente adequadas para processar dados sequenciais, como texto, fala e séries temporais. Existem vários tipos de RNNs, cada um com suas próprias características e aplicações. Aqui estão alguns dos tipos mais comuns:

**1. RNN Simples (Vanilla RNN):**

* **Estrutura:** A RNN simples é a estrutura básica de uma RNN. Consiste em uma única camada que recebe a entrada e um estado oculto. O estado oculto é atualizado em cada passo de tempo, levando em consideração a entrada atual e o estado oculto anterior.
* **Vantagens:**  Simples e fácil de entender.
* **Desvantagens:**  Luta para aprender dependências de longo prazo.

**2. RNN com Memória de Longo Prazo (LSTM - Long Short-Term Memory):**

* **Estrutura:**  O LSTM resolve o problema de dependências de longo prazo da RNN simples, introduzindo portas que controlam o fluxo de informações. As portas são usadas para:
    * Esquecer informações irrelevantes do passado.
    * Armazenar informações importantes no estado oculto.
    * Controlar o fluxo de informações para a saída.
* **Vantagens:**  Pode aprender dependências de longo prazo em dados sequenciais.
* **Desvantagens:**  Mais complexo e computacionalmente mais caro que a RNN simples.

**3. RNN com Unidades de Porta Recorrente (GRU - Gated Recurrent Unit):**

* **Estrutura:** O GRU é uma variação simplificada do LSTM.  Ele combina as portas de "esquecer" e "entrada" em uma única porta.
* **Vantagens:**  Menos parâmetros e mais eficiente computacionalmente que o LSTM.
* **Desvantagens:**  Pode não ser tão poderoso como o LSTM para aprender dependências de longo prazo em alguns casos.

**4. RNN Bidirecional (BiRNN):**

* **Estrutura:** O BiRNN processa a sequência de entrada em ambas as direções (para frente e para trás). Isso permite que a rede capte contexto de ambas as extremidades da sequência.
* **Vantagens:**  Pode capturar dependências em ambas as direções da sequência.
* **Desvantagens:**  Mais complexo e computacionalmente mais caro que uma RNN simples.

**5. Redes Neurais Recorrentes Convolucionais (ConvRNN):**

* **Estrutura:** Combina convoluções com RNNs para lidar com dados sequenciais e espaciais. As convoluções são aplicadas nas entradas temporais, enquanto a RNN processa as saídas das convoluções.
* **Vantagens:**  Pode capturar características espaciais e temporais nos dados.
* **Desvantagens:**  Pode ser mais complexo de implementar que RNNs simples.


**Diferenças:**

| Tipo de RNN | Descrição | Vantagens | Desvantagens |
|---|---|---|---|
| RNN Simples | Estrutura básica da RNN. | Simples e fácil de entender. |  Luta para aprender dependências de longo prazo. |
| LSTM | Possui portas para controlar o fluxo de informações, resolvendo o problema de dependências de longo prazo. |  Pode aprender dependências de longo prazo. |  Mais complexo e computacionalmente mais caro. |
| GRU | Versão simplificada do LSTM. |  Menos parâmetros e mais eficiente. |  Pode não ser tão poderoso quanto o LSTM em alguns casos. |
| BiRNN | Processa a sequência em ambas as direções. |  Captura dependências em ambas as direções. |  Mais complexo e computacionalmente mais caro. |
| ConvRNN | Combina convoluções com RNNs. |  Captura características espaciais e temporais. |  Pode ser mais complexo de implementar. |

**Escolha do tipo de RNN:**

A escolha do tipo de RNN depende da tarefa específica e dos dados. Para tarefas simples que não requerem o aprendizado de dependências de longo prazo, a RNN simples pode ser suficiente. Para tarefas mais complexas com dependências de longo prazo, o LSTM ou o GRU podem ser melhores opções. O BiRNN pode ser útil quando o contexto da sequência é importante em ambas as direções. O ConvRNN é apropriado para dados sequenciais com componentes espaciais.

## O que são Dados Sequenciais?

Dados sequenciais são dados—como palavras, frases, áudio (fala) ou dados de séries temporais—onde os componentes sequenciais estão inter-relacionados com base em regras complexas de semântica e sintaxe.

## Quais são os tipos de redes neurais recorrentes?

A seguir, estão alguns tipos comuns de RNNs.

* **RNNs Um-para-Muitos**: Nesta arquitetura, a RNN recebe uma única entrada no início (a parte "um") e, em seguida, gera uma sequência de saídas (a parte "muitos").

   - Isso pode ser útil para tarefas como legendagem de imagens, onde a entrada é uma imagem e a saída é uma sequência de palavras descrevendo o conteúdo da imagem.
   - **Exemplo**: *Na legendagem de imagens, a RNN recebe uma imagem como entrada e gera uma sequência de palavras para descrever o conteúdo da imagem.*

* **RNNs Muitos-para-Um**: Por outro lado, nas RNNs muitos-para-um, a rede recebe uma sequência de entradas (a parte "muitos") e produz uma única saída (a parte "um").
   - Esta configuração é comumente usada para tarefas como análise de sentimentos, onde a entrada é uma sequência de palavras e a saída é um rótulo de sentimento (positivo ou negativo).
   - **Exemplo**: *Na análise de sentimentos, a RNN recebe uma sequência de palavras representando uma crítica de filme e prevê se o sentimento é positivo, negativo ou neutro a partir das críticas de entrada.*

* **RNNs Muitos-para-Muitos (Sequência-para-Sequência)**: Nesta arquitetura, a RNN recebe uma sequência de entradas e produz uma sequência de saídas, podendo ser de comprimentos variáveis.

   - Esta configuração é usada para tarefas como tradução automática, onde a entrada é uma sequência de palavras em um idioma e a saída é uma sequência de palavras em outro idioma.
   - **Exemplo**: *Na tradução automática, a RNN recebe uma sequência de palavras em inglês e gera uma sequência de palavras em francês.*

* **RNNs Muitos-para-Muitos (Sincronizados)**: Nesta variante das RNNs muitos-para-muitos, as sequências de entrada e saída têm o mesmo comprimento.

   - Esta arquitetura é usada para tarefas como reconhecimento de entidades nomeadas (NER), onde a entrada é uma sequência de palavras e a saída é uma sequência de rótulos indicando entidades nomeadas.
   - **Exemplo**: *No reconhecimento de entidades nomeadas, a RNN recebe uma sequência de palavras e gera uma sequência de rótulos indicando se cada palavra faz parte de uma entidade nomeada (por exemplo, pessoa, organização, localização).*

### E quanto ao Um-para-Um?

Embora a arquitetura "um-para-um" seja direta (cada entrada é processada independentemente, e não há conexões recorrentes ou loops de feedback) e adequada para certos tipos de tarefas, ela não aproveita a natureza sequencial dos dados nem captura dependências temporais, que são essenciais para tarefas como previsão de sequência, previsão de séries temporais ou processamento de linguagem natural.

   - Esta arquitetura é usada para tarefas como tarefas simples de classificação, tarefas de regressão ou tarefas onde cada entrada representa uma instância independente.

   - **Exemplo**: *Suponha que você tenha um conjunto de dados de imagens, e cada imagem esteja associada a um único rótulo indicando o objeto na imagem. Você pode treinar uma rede neural "um-para-um" para classificar cada imagem em sua respectiva categoria sem considerar qualquer informação sequencial ou temporal.*

## Como Funciona o Modelo RNN?

Ao contrário das redes neurais feedforward, que processam cada entrada de forma independente,

as RNNs mantêm um **estado interno (memória)** que lhes permite *capturar dependências temporais e padrões em dados sequenciais*.

* **Entrada Sequencial**: As RNNs recebem dados de entrada sequenciais, como texto, séries temporais ou áudio, onde a ordem dos elementos é importante. Cada elemento na sequência é tipicamente representado como um vetor ou uma sequência de vetores de características.

* **Conexões Recorrentes**: As RNNs têm conexões recorrentes que permitem que a informação persista ao longo dos passos de tempo. Em cada passo de tempo, a RNN processa um vetor de entrada e atualiza seu estado interno com base tanto na entrada atual quanto no estado anterior.

* **Atualização do Estado Oculto**: Em cada passo de tempo \( t \), a RNN calcula um novo estado oculto \( h_t \) usando a entrada atual \( x_t \) e o estado oculto anterior \( h_{t-1} \). Este estado oculto serve como um resumo da informação processada até o momento e é passado para o próximo passo de tempo.

## **ℎ 𝑡 = 𝑓(𝑊 ℎ ⋅ ℎ 𝑡−1 + 𝑊 𝑥 ⋅𝑥 𝑡 + 𝑏)**


where:

* ℎ
𝑡 is the hidden state at time step 𝑡

* 𝑓 is the activation function (e.g., tanh or ReLU)

* 𝑊
ℎ
is the weight matrix for the recurrent connections

* 𝑊
𝑥
 is the weight matrix for the input connections

* 𝑏 is the bias vector

* **Saída**: Dependendo da tarefa, a RNN pode produzir uma saída em cada passo de tempo (sequência-para-sequência) ou apenas no passo de tempo final (sequência-para-vetor). A saída em cada passo de tempo \( t \) é calculada com base no estado oculto atual \( h_t \).

* **Treinamento**: As RNNs são tipicamente treinadas usando backpropagation through time (BPTT), que é uma extensão do algoritmo de backpropagation padrão. O BPTT calcula os gradientes da função de perda em relação aos parâmetros do modelo (pesos e biases) desenrolando a rede ao longo do tempo e aplicando a regra da cadeia.

```python
model = Sequential()
model.add(Input(shape=(100,1)))
model.add(SimpleRNN(32, return_sequences=False))
model.add(Dense(1,activation='sigmoid'))

model.summary()
```
